```python
def convert():
    for i in tqdm(range(1, 10)):
        data = pd.read_json(f'../data/json/{i}.json', lines = True)
        data.to_csv(f'../data/csv/{i}.csv')
```

```python
def get_cols_name():
    cols = set()
    for i in tqdm(range(1, 10)):
        data = pd.read_csv(f'../data/csv/{i}.csv', low_memory = False)
        cols.update(data.columns.values.tolist())
    return cols
```
